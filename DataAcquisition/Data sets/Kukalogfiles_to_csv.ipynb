```python
import pandas as pd
import io
import glob
```

Read text files first

Insert folder name containing the dataset 

```python
datafolder =r'13may2023' 
path = datafolder + r'/*.log'
files = glob.glob (path) #find all file names inside the directory with the type .log
print(files)
print(len(files))
```

```python
# group files based on their main specs : applied force , collision type
# 'NoColl' means the data recorcded when robot moves along the desired path(Force and pose) and no collision occurs (collision free)
logdata10N = {}
logdata10N["NoColl"] =[fileName for fileName in files if fileName.find("f10") != -1 and fileName.find("NoColl") != -1]   
print(len(logdata10N["NoColl"]))  # shows how many .log file for 10N force and No collision is available

logdata10N["Intentional"] =[fileName for fileName in files if fileName.find("f10") != -1 and fileName.find("Intentional") != -1]   
print(len(logdata10N["Intentional"]))

logdata10N["Accidental"] =[fileName for fileName in files if fileName.find("f10") != -1 and fileName.find("Accidntal") != -1]   
print(len(logdata10N["Accidental"]))

logdata15N = {}
logdata15N["NoColl"] = [fileName for fileName in files if fileName.find("f15") != -1 and fileName.find("NoColl") != -1]  
print(len(logdata15N["NoColl"]))

logdata15N["Intentional"] = [fileName for fileName in files if fileName.find("f15") != -1 and fileName.find("Intentional") != -1]   
print(len(logdata15N["Intentional"]))

logdata15N["Accidental"] = [fileName for fileName in files if fileName.find("f15") != -1 and fileName.find("Accidntal") != -1]   # there is a typo in Accidntal, fix in kuka
print(len(logdata15N["Accidental"]))

```

```python
dataCircularPath_f10N = pd.DataFrame()
dataCircularPath_f15N = pd.DataFrame()


for logfile in logdata10N["NoColl"]:
    with open(logfile, 'r') as f:
        file_str = f.read()
        # Removing the character '%' that KUKA generates in the beginning of the log files
        file_str2 = file_str[1:]
        data = pd.read_csv (io.StringIO(file_str2),delim_whitespace=True)
        data['CollisionType'] = 'NoColl'
        dataCircularPath_f10N = pd.concat([dataCircularPath_f10N,data],ignore_index = True)
        print(dataCircularPath_f10N.shape)  

for logfile in logdata10N["Intentional"]:
    with open(logfile, 'r') as f:
        file_str = f.read()
        file_str2 = file_str[1:]
        data = pd.read_csv (io.StringIO(file_str2),delim_whitespace=True)
        data['CollisionType'] = 'Intentional'
        dataCircularPath_f10N = pd.concat([dataCircularPath_f10N,data],ignore_index = True)
        print(dataCircularPath_f10N.shape)  

for logfile in logdata15N["NoColl"]:
    with open(logfile, 'r') as f:
        file_str = f.read()
        file_str2 = file_str[1:]
        data = pd.read_csv (io.StringIO(file_str2),delim_whitespace=True)
        data['CollisionType'] = 'NoColl'
        dataCircularPath_f15N = pd.concat([dataCircularPath_f15N,data],ignore_index = True)
        print(dataCircularPath_f15N.shape)   

for logfile in logdata15N["Intentional"]:
    with open(logfile, 'r') as f:
        file_str = f.read()
        file_str2 = file_str[1:]
        data = pd.read_csv (io.StringIO(file_str2),delim_whitespace=True)
        data['CollisionType'] = 'Intentional'
        dataCircularPath_f15N = pd.concat([dataCircularPath_f15N,data],ignore_index = True)
        print(dataCircularPath_f15N.shape) 

for logfile in logdata15N["Accidental"]:
    with open(logfile, 'r') as f:
        file_str = f.read()
        file_str2 = file_str[1:]
        data = pd.read_csv (io.StringIO(file_str2),delim_whitespace=True)
        data['CollisionType'] = 'Accidental'
        dataCircularPath_f15N = pd.concat([dataCircularPath_f15N,data],ignore_index = True)
        print(dataCircularPath_f15N.shape)  

dataCircularPath_f10N.columns = dataCircularPath_f15N.columns = data.columns.values
```
